# The for Keyword in Python
The for keyword in Python is used to create loops that iterate over sequences (iterables). It's one of the most fundamental tools for repetitive tasks and processing collections of data.

# Basic Syntax
```python
for variable in iterable:
    # Code block to repeat for each item
    print(variable)
```
# How It Works
* Takes each item from the iterable one at a time
* Assigns the item to the variable
* Executes the code block for each item
* Repeats until all items are processed

# Common Iterables
| Iterable | Example |
|----------|---------|
| `list` | `[1, 2, 3]` |
| `tuple` | `(1, 2, 3)` |
| `string` | `"Python"(iterates over characters)`|
| `range` | `range(5)` |
| `set` | `{1, 2, 3}` |
| `dictionary` | `{"a": 1, "b": 2}` *(iterates over keys by default)* |

# Common Patterns
| Task | Pattern |
|------|---------|
| Loop `n` times | `for i in range(n):` |
| Loop with index | `for i, item in enumerate(iterable):` |
| Loop over a dictionary | `for key, value in dict.items():` |
| Loop backwards | `for i in range(n-1, -1, -1):` |
| Iterate over two lists together | `for a, b in zip(list1, list2):` |

# Summary Table
| Concept | Description |
|---------|-------------|
| Purpose | Iterate over the elements of a sequence or other iterable. |
| Common with | `list`, `tuple`, `string`, `range()`, `set`, and `dict` |
| Flow control | Supports `break`, `continue`, and `else` |
| Alternatives | `while` loops, list comprehensions, generator expressions |
| Performance | Generally fast and efficient for iterating over iterables in Python |



# Examples

## 1. Iterating Over a List

````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
# Output:
# apple
# banana
# cherry
````

## 2. Iterating Over a String

````python
word = "Python"
for letter in word:
    print(letter)
# Output:
# P
# y
# t
# h
# o
# n
````

## 3. Using range() for Numeric Loops

````python
# Print numbers 0 to 4
for i in range(5):
    print(i)
# Output: 0, 1, 2, 3, 4

# Print numbers 2 to 8 (step 2)
for i in range(2, 10, 2):
    print(i)
# Output: 2, 4, 6, 8
````

## 4. Iterating Over Dictionaries

````python
person = {"name": "Alice", "age": 30, "city": "New York"}

# Iterate over keys
for key in person:
    print(f"{key}: {person[key]}")

# Or using items()
for key, value in person.items():
    print(f"{key}: {value}")
# Output:
# name: Alice
# age: 30
# city: New York
````

# Advanced Patterns

## 1. Nested for Loops

````python
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
# Output:
# i=0, j=0
# i=0, j=1
# i=1, j=0
# i=1, j=1
# i=2, j=0
# i=2, j=1
````

## 2. Looping with enumerate() (Index + Value)

````python
colors = ["red", "green", "blue"]
for index, color in enumerate(colors):
    print(f"{index}: {color}")
# Output:
# 0: red
# 1: green
# 2: blue
````

## 3. Looping with zip() (Multiple Iterables)

````python
names = ["Alice", "Bob", "Charlie"]
scores = [85, 92, 78]

for name, score in zip(names, scores):
    print(f"{name} scored {score}")
# Output:
# Alice scored 85
# Bob scored 92
# Charlie scored 78
````

## 4. List Comprehension (Compact for Loop)

````python
# Traditional loop
squares = []
for x in range(5):
    squares.append(x**2)

# List comprehension (more Pythonic)
squares = [x**2 for x in range(5)]
print(squares)  # Output: [0, 1, 4, 9, 16]
````

# Controlling Loop Flow

## break - Exit Loop Early

````python
for num in range(10):
    if num == 5:
        break  # Stops at 5
    print(num)
# Output: 0, 1, 2, 3, 4
````

## continue - Skip Current Iteration

````python
for num in range(5):
    if num == 2:
        continue  # Skips 2
    print(num)
# Output: 0, 1, 3, 4
````

## else with for (Runs if no break)

````python
for num in range(5):
    if num == 10:
        break
else:
    print("Loop completed without break!")  # ✅ Runs
# Output: Loop completed without break!
````

# Quick Reference

```python
# Basic loop
for item in iterable:
    process(item)

# With index
for i, item in enumerate(iterable):
    print(f"{i}: {item}")

# Numeric range
for i in range(start, stop, step):
    print(i)

# Dictionary
for key, value in dict.items():
    print(key, value)

# Multiple iterables
for a, b in zip(list1, list2):
    print(a, b)

# Conditional
for item in iterable:
    if condition:
        break  # or continue

# Else clause (if no break)
for item in iterable:
    if condition:
        break
else:
    print("No break occurred")
```
