# The finally Keyword in Python
The finally keyword is used in exception handling to define a block of code that always executes, regardless of whether an exception occurred or was caught. It's typically used for cleanup actions like closing files, releasing resources, or disconnecting from databases.

# Basic Syntax
```python
try:
    # Code that might raise an exception
    risky_operation()
except SomeException:
    # Handle the exception
    print("Error handled")
finally:
    # This ALWAYS runs, no matter what
    print("Cleanup complete")
```

# How It Works
| Scenario | Does `finally` run? |
|----------|----------------------|
| No exception occurs | ✅ Yes |
| Exception occurs and is caught | ✅ Yes |
| Exception occurs and is **not** caught | ✅ Yes (then the exception propagates) |
| `return` statement in `try` or `except` | ✅ Yes (executes before returning) |
| `break` or `continue` in a loop | ✅ Yes (executes before breaking or continuing) |

# Important Notes
* finally is optional - you don't always need it
* Always runs - even if return, break, or continue is used
* Can be used without except - for guaranteed cleanup
* Exceptions in finally - can mask original exceptions (use carefully!)
* Use for cleanup - closing files, releasing locks, etc.

# Summary
 Feature | `finally` |
|---------|-----------|
| Purpose | Always execute cleanup code |
| Runs? | Yes, regardless of whether an exception occurs or not |
| Common uses | Closing files, disconnecting from databases, releasing locks, cleaning up resources |
| Syntax | Placed after the `try`, `except`, and optional `else` blocks |
| Optional? | Yes (but recommended when resource cleanup is required) |

# Examples

## 1. Basic Cleanup (Always Runs)

````python
try:
    x = 10 / 2
    print(f"Result: {x}")
except ZeroDivisionError:
    print("Cannot divide by zero!")
finally:
    print("This ALWAYS runs!")  # ✅ Always runs

# Output:
# Result: 5.0
# This ALWAYS runs!
````

## 2. Even When Exception Occurs

````python
try:
    x = 10 / 0  # 💥 Raises exception
    print("This won't run")
except ZeroDivisionError:
    print("Error caught!")
finally:
    print("Cleanup runs even with error!")  # ✅ Still runs

# Output:
# Error caught!
# Cleanup runs even with error!
````

## 3. With return in try (finally runs first)

````python
def test_finally():
    try:
        return "Returning from try"
    finally:
        print("Finally runs BEFORE return!")  # ✅ Runs first
    print("This never runs")

result = test_finally()
print(f"Result: {result}")

# Output:
# Finally runs BEFORE return!
# Result: Returning from try
````

# Real-World Use Cases

## 1. File Handling (Ensure File is Closed)

````python
def read_file(filename):
    file = None
    try:
        file = open(filename, 'r')
        content = file.read()
        return content
    except FileNotFoundError:
        print(f"File {filename} not found!")
        return None
    finally:
        if file:
            file.close()  # ✅ Always closes the file
            print("File closed!")

print(read_file("./data.txt"))
````

## 2. Database Connections

````python
import sqlite3

def get_user_data(user_id):
    conn = None
    cursor = None
    try:
        conn = sqlite3.connect('database.db')
        cursor = conn.cursor()
        cursor.execute("SELECT * FROM users WHERE id = ?", (user_id,))
        return cursor.fetchone()
    except sqlite3.Error as e:
        print(f"Database error: {e}")
        return None
    finally:
        if cursor:
            cursor.close()  # ✅ Always close cursor
        if conn:
            conn.close()    # ✅ Always close connection
            print("Database connection closed!")
````

## 3. Network/API Calls

````python
import requests

def fetch_api_data(url):
    response = None
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()
        return response.json()
    except requests.RequestException as e:
        print(f"API error: {e}")
        return None
    finally:
        if response:
            response.close()  # ✅ Always close connection
            print("Connection closed!")
````

## 4. Lock/Resource Release

````python
import threading

lock = threading.Lock()

def thread_safe_operation():
    try:
        lock.acquire()
        print("Resource acquired, doing work...")
        # Risky code here
    except Exception as e:
        print(f"Error: {e}")
    finally:
        lock.release()  # ✅ Always releases lock
        print("Resource released!")
````

# finally with else and except
## Complete Exception Handling Structure

````python
try:
    x = 10 / 2
    print("Try block executed")
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print(f"Success! Result: {x}")  # ✅ Runs only if no exception
finally:
    print("Cleanup: ALWAYS runs!")  # ✅ Always runs

# Output:
# Try block executed
# Success! Result: 5.0
# Cleanup: ALWAYS runs!
````

## Order of Execution

````python
try:
    print("1. Try block")
    # Code that might raise error
except:
    print("2. Except block (if error)")
else:
    print("2. Else block (if no error)")
finally:
    print("3. Finally block ALWAYS")
````
