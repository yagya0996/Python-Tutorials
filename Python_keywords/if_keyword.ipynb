# The if Keyword in Python
The if keyword is a conditional statement that executes a block of code only if a specified condition evaluates to True. It's the foundation of decision-making in Python.

# Basic Syntax
```python
if condition:
    # Executes only if condition is True
    code_block
```
# Key Concepts
1. **Condition Evaluation**
    * Any expression that returns True or False
    * Truthy values execute (non-zero numbers, non-empty strings/lists, etc.)
    * Falsy values skip execution (False, None, 0, "", [], {}, ())

2. **The Full Structure**
    ```python
    if condition1:
        # Runs if condition1 is True
    elif condition2:
        # Runs if condition1 is False AND condition2 is True
    elif condition3:
        # Runs if previous are False AND condition3 is True
    else:
        # Runs if ALL previous conditions are False
    ```
3. **Key Points**
    * elif = "else if" - checks additional conditions
    * else = "catch-all" - runs when no conditions are True
    * Only one block executes (mutually exclusive)
    * Conditions are checked top to bottom, stops at first True
    * Indentation matters (4 spaces recommended)

Summary
| Concept |	Description |
|---------|-------------|
| if	| Executes block if condition is True |
| elif	| Checks additional conditions (optional) |
| else  | Catch-all (optional) |
| indentation |	Defines which code belongs to the if |



# Examples

````python
age = 18

if age < 13:
    status = "Child"
elif age < 20:
    status = "Teenager"   # ✅ Runs
elif age < 65:
    status = "Adult"
else:
    status = "Senior"

print(status)  # Output: Teenager
````

````python
# Single-line if (valid but less readable)
x = 10
if x > 5: print("Greater")  # ✅ Works

# Ternary operator (conditional expression)
result = "Even" if x % 2 == 0 else "Odd"
````

# Common Pitfalls
```python
# ❌ Wrong - assignment in condition
if x = 5:  # SyntaxError

# ❌ Wrong - missing colon
if x > 5   # SyntaxError

# ❌ Wrong - incorrect indentation
if x > 5:
print("Hello")  # IndentationError
```
