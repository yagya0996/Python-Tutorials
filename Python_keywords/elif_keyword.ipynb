# The elif Keyword in Python
The elif keyword stands for "else if" and is used in conditional statements to check multiple conditions after an initial if statement. It allows you to test several conditions in sequence, executing the first block where the condition evaluates to True.

# Basic Syntax
```python
if condition1:
    # Code block executes if condition1 is True
elif condition2:
    # Code block executes if condition1 is False AND condition2 is True
elif condition3:
    # Code block executes if condition1 and condition2 are False AND condition3 is True
else:
    # Code block executes if ALL previous conditions are False
```

# How It Works
    1. Python checks conditions from top to bottom
    2. The first True condition triggers its block and skips all remaining conditions
    3. If no conditions are True, the else block runs (if present)

# Key Differences from Nested if	
| Approach              | Behavior                                           |
|-----------------------|----------------------------------------------------|
| `if` + `elif`         | Mutually exclusive – only **one** block runs.      |
| Nested `if` statements| Multiple blocks can run if all conditions are `True`. |

# Important Notes
    1. You can have multiple elif statements (unlimited)
    2. else is optional (you don't always need a final catch-all)
    3. Indentation matters - elif must be at the same indentation level as if
    4. elif cannot appear without a preceding if (syntax error)
    5. Order matters - put the most specific conditions first

# When to Use elif?
    1. When you have multiple exclusive conditions
    2. For decision trees (like grading, menus, categorization)
    3. To make code cleaner and more readable than multiple nested if statements

## Examples

## 1. Grade Calculator

````python
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"      # ✅ This runs (85 >= 80)
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Grade: {grade}")  # Output: Grade: B
````

## 2. Multiple Conditions with elif

````python
age = 25

if age < 13:
    print("Child")
elif age < 20:
    print("Teenager")
elif age < 65:
    print("Adult")    # ✅ This runs (25 < 65)
else:
    print("Senior")
````

## 3. Key Differences from Nested if

````python
# Using elif (only one runs)
x = 10
if x > 5:
    print("Greater than 5")   # ✅ Runs
elif x > 3:
    print("Greater than 3")   # ❌ Skipped

# Using separate if statements (both can run)
if x > 5:
    print("Greater than 5")   # ✅ Runs
if x > 3:
    print("Greater than 3")   # ✅ Also runs
````

## 4. User Input Validation

````python
user_input = input("Enter 'yes', 'no', or 'maybe': ")

if user_input == "yes":
    print("You chose yes!")
elif user_input == "no":
    print("You chose no!")
elif user_input == "maybe":
    print("You chose maybe!")
else:
    print("Invalid input!")
````

## 5. Menu Selection

````python
choice = 2

if choice == 1:
    print("Option 1 selected")
elif choice == 2:
    print("Option 2 selected")  # ✅ Runs
elif choice == 3:
    print("Option 3 selected")
else:
    print("Invalid option")
````

## Common Mistakes

### ❌ Wrong:

````python
elif x == 5:  # SyntaxError - no matching 'if'
    print("x is 5")
````

### ✅ Correct:

````python
x=0
if x > 0:
    print("Positive")
elif x == 0:
    print("Zero")
else:
    print("Negative")
````
