# The global Keyword in Python
The global keyword is used to declare that a variable inside a function refers to a variable defined in the global (top-level) scope, rather than creating a new local variable. It allows you to modify global variables from within a function.

# Basic Syntax
```python
global variable_name
```

# Why We Need global
By default, variables assigned inside a function are local to that function. Without global, you can read global variables, but if you try to assign a value to them, Python creates a new local variable instead.

# When to Use (and Avoid) global
✅ Good Uses
* Configuration/Settings (app-wide settings)
* Counting (number of function calls)
* Caching (store computed results)
* State management in simple scripts
* Game variables (score, lives, level)

❌ Bad Uses (Avoid)
* Large projects with many functions
* Instead of passing parameters (functions should be self-contained)
* Constants (use uppercase variable names instead)
* When functions get complex (use classes instead)

# Summary Table
| Aspect | Description |
|--------|-------------|
| Purpose | Modify global (module-level) variables inside functions. |
| Scope | Affects variables defined at the module level. |
| Alternatives | Pass parameters, return values, or use classes/objects. |
| Best practice | Use sparingly; prefer function arguments and return values when possible. |
| Common use | Configuration settings, counters, flags, and simple scripts. |

# global vs nonlocal
|Keyword | Scope | When to use |
|--------|-------|-------------|
|global  | Global (top-level) scope | Modify variables defined at module level |
|nonlocal|	Enclosing (outer) function scope | Modify variables in nested functions |



# Examples

## 1. Reading Global Variables (No global Needed)

````python
x = 10  # Global variable

def read_global():
    print(x)  # ✅ Can read global variables without 'global'

read_global()  # Output: 10
````

## 2. Assigning to Global Variables (Need global)

````python
x = 10  # Global variable

def modify_global():
    x = 20  # ❌ This creates a NEW local variable 'x'
    print(x)

modify_global()  # Output: 20
print(x)         # Output: 10 (global x is unchanged!)
````

## 3. Using global to Modify Global Variables

````python
x = 10  # Global variable

def modify_global():
    global x  # ✅ Declare that we're using the global 'x'
    x = 20    # ✅ Modifies the global variable

modify_global()
print(x)  # Output: 20 (global x is changed!)
````

# Common Mistakes and Pitfalls

## ❌ Forgetting global when reassigning

````python
count = 5

def increment():
    count = count + 1  # ❌ UnboundLocalError!
    print(count)
    # Python sees an assignment, so it thinks 'count' is local
    # But local 'count' is used before assignment (RHS needs it)
increment()
````

## Fix:

````python
count = 5

def increment():
    global count
    count = count + 1  # ✅ Works!
    print(count)

increment()
````

## ❌ Using global on a variable that doesn't exist

````python
def create_global():
    global new_var
    new_var = 100  # Creates a new global variable

create_global()
print(new_var)  # Output: 100 (global variable created!)
````

# Quick Reference

````python
# Declare a global variable
x = 10

def function():
    global x
    # Access (read) global - no need for global
    print(x)  # ✅ Works
    
    # Modify global - NEED global
    
    x = 20  # ✅ Modifies global
    
    # Create new global from inside function
    global new_var
    new_var = 30  # ✅ Creates global variable

function()
print(x)       # Output: 20
print(new_var) # Output: 30
````
