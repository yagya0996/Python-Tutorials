# The except Keyword in Python
The except keyword is used in Python for exception handling – it catches and handles errors (exceptions) that occur in the try block, preventing your program from crashing.

# Basic Syntax
```python
try:
    # Code that might raise an exception
    risky_operation()
except:
    # Code that runs if ANY exception occurs
    handle_error()
```

# How It Works
1. Python executes the code inside the try block
2. If no exception occurs, the except block is skipped
3. If an exception occurs, Python jumps to the except block
4. The program continues running (doesn't crash)

# When Each Block Runs
| Block | When It Runs |
|-------|--------------|
| `try` | Always executes first (unless the code flow never reaches it). |
| `except` | Executes only if an exception occurs in the `try` block. |
| `else` | Executes only if **no exception** occurs in the `try` block. |
| `finally` | Always executes, whether or not an exception occurs or is handled. |

# Key Notes
1. Be Specific: Catch specific exceptions rather than using a bare except:
2. Order Matters: Put more specific exceptions before broader ones
3. Don't Swallow Exceptions: Always log or handle errors meaningfully
4. except Exception catches most errors, but not all (e.g., KeyboardInterrupt)

# Common Exception Types
| Exception | Description |
|-----------|-------------|
| `ZeroDivisionError` | Raised when attempting to divide by zero. |
| `ValueError` | Raised when a function receives a valid type but an invalid value. |
| `TypeError` | Raised when an operation or function is applied to an object of an inappropriate type. |
| `FileNotFoundError` | Raised when a file or directory cannot be found. |
| `IndexError` | Raised when a list or sequence index is out of range. |
| `KeyError` | Raised when a dictionary key is not found. |
| `AttributeError` | Raised when an object does not have the requested attribute. |

# Summary
* except catches and handles exceptions
* Always use try before except
* Catch specific exceptions whenever possible
* Use else for code that runs when no exception occurs
* Use finally for cleanup code that must run

# Examples

# 1. Catching Any Exception (Not Recommended)

````python
try:
    x = 10 / 0  # This raises ZeroDivisionError
    print("This won't run")
except:
    print("An error occurred!")  # ✅ Runs
````

# 2. Catching Specific Exceptions (Recommended)

````python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")  # ✅ Runs

print("Program continues...")  # ✅ Still runs
````

# 3. Catching Multiple Exception Types

````python
try:
    user_input = input("Enter a number: ")
    num = int(user_input)          # Could raise ValueError
    result = 10 / num              # Could raise ZeroDivisionError
    
except ValueError:
    print("That's not a valid number!")
except ZeroDivisionError:
    print("Cannot divide by zero!")
````

# 4. Catching Multiple Exceptions in One Block

````python
try:
    num = int(input("Enter a number: "))
    result = 100 / num
except (ValueError, ZeroDivisionError):
    print("Invalid input or division by zero!")
````

# 5. Accessing the Exception Object
Use 'as' to capture the exception details:

````python
try:
    x = 10 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")  # Output: Error: division by zero
````

# 6. Catching All Exceptions (Use Sparingly)

````python
try:
    risky_code()
except Exception as e:  # Catches ALL built-in exceptions
    print(f"Something went wrong: {e}")
````

# Advanced: except with else and finally
## The Full try/except/else/finally Structure

````python
try:
    x = 10 / 2
    print("Try block executed")
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print(f"Success! Result: {x}")  # ✅ Runs only if no exception
finally:
    print("This ALWAYS runs")  # ✅ Runs regardless of exceptions
````

# Real-World Examples

## Example 1: Safe File Reading

````python
try:
    with open("data.txt", "r") as file:
        content = file.read()
except FileNotFoundError:
    print("File not found! Creating a new one...")
    with open("data.txt", "w") as file:
        file.write("Default content")
except PermissionError:
    print("Don't have permission to read the file!")
````

## Example 2: Input Validation

````python
def get_valid_number(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("Please enter a valid integer!")

age = get_valid_number("Enter your age: ")
print(f"Your age is {age}")
````

## Example 3: Handling Network Errors

````python
import requests

try:
    response = requests.get("https://api.example.com/data", timeout=5)
    data = response.json()
except requests.ConnectionError:
    print("Network error - check your connection")
except requests.Timeout:
    print("Request timed out - server might be slow")
except requests.JSONDecodeError:
    print("Invalid JSON response from server")
except requests.RequestException as e:
    print(f"Something else went wrong: {e}")
````
