# The from Keyword in Python
The from keyword in Python is used for importing specific attributes, functions, classes, or modules from a module or package. It allows you to selectively bring parts of a module into your current namespace, rather than importing the entire module.

# Basic Syntax
```python
from module import item
from module import item1, item2, item3
from module import item as alias
from package.submodule import item
```
# How It Works
1. Specifies the source (module/package) to import from
2. Imports only the specified items into the current namespace
3. Allows direct usage without prefixing the module name

# Summary Table

| Use Case | Example |
|----------|---------|
| Specific functions | `from math import sqrt, pi` |
| Class import | `from datetime import datetime` |
| With alias | `from json import loads as load_json` |
| Multiple items | `from os.path import join, exists, basename` |
| From a package | `from django.db import models` |
| Wildcard *(not recommended)* | `from math import *` |


# Common Use Cases

## 1. Importing Specific Functions/Classes

````python
# Instead of:
import math
print(math.sqrt(16))  # Need to prefix with 'math.'

# Use from:
from math import sqrt
print(sqrt(16))  # Direct usage
````

## 2. Importing Multiple Items

````python
from math import sqrt, pi, sin, cos

print(sqrt(25))     # Output: 5.0
print(pi)           # Output: 3.14159...
print(sin(0))       # Output: 0.0
````

## 3. Importing with Aliases (Renaming)

````python
from datetime import datetime as dt

now = dt.now()
print(now)  # Current date/time
````

## 4. Importing All Items (Not Recommended)

````python
from math import *  # Imports everything (bad practice)

print(sqrt(25))     # Works but you lose track of origins
print(pi)
# ⚠️ Danger: Can cause naming conflicts!
````

# Importing from Packages

## 1. From Submodules

````python
# Import from a submodule
from os.path import join, exists

path = join("folder", "file.txt")
print(path)
if exists(path):
    print("File exists!")
````

## 2. From Nested Packages

````python
from django.contrib.auth import get_user_model

User = get_user_model()  # Works directly
````

# Importing from Local Files
Example Project Structure
```text
project/
├── main.py
└── utils/
    ├── __init__.py
    └── helpers.py
```

## From helpers.py

````python
# helpers.py
def greet(name):
    return f"Hello, {name}!"

def calculate_sum(a, b):
    return a + b
````

## In main.py

````python
# Option 1: Import specific functions
from utils.helpers import greet, calculate_sum

print(greet("Alice"))          # Output: Hello, Alice!
print(calculate_sum(5, 3))     # Output: 8

# Option 2: Import with alias
from utils.helpers import greet as say_hello

print(say_hello("Bob"))        # Output: Hello, Bob!
````

# Importing __all__ from a Package

````python
# In package __init__.py
__all__ = ['module1', 'module2']

# Then you can do:
from package import *  # Imports only modules in __all__
````
