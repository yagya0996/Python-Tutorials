# The del Keyword in Python

The del keyword in Python is used to delete objects, including variables, items from lists, keys from dictionaries, and attributes from objects. It's a way to explicitly remove references to objects in Python.

# Important Notes
    1. Memory Management: del removes the reference to an object, but the object itself is only deleted if no other references exist (via Python's garbage collection).
    2. What It Doesn't Do:
        a. Doesn't immediately free memory (Python handles this automatically)
        b. Can't delete individual characters from strings (strings are immutable)
    3. Common Use Cases:
        a. Freeing up memory by removing large objects
        b. Cleaning up namespaces
        c. Removing specific elements from mutable collections

# When to Use del
    1. When you need to explicitly remove a reference
    2. When working with large data structures and want to free memory
    3. When cleaning up names in a namespace
    4. Remember that in most cases, Python's automatic garbage collection will handle memory management for you, so explicit use of del isn't always necessary.

# Basic Usage

## 1. Deleting Variables

```python
x = 10
print(x)  # Output: 10
del x
print(x)  # Raises NameError: name 'x' is not defined
```

## 2. Deleting List Elements

```python
my_list = [1, 2, 3, 4, 5]
del my_list[1]    # Delete the second element
print(my_list)    # Output: [1, 3, 4, 5]

del my_list[1:3]  # Delete slice (elements 1 and 2)
print(my_list)    # Output: [1, 5]
```

## 3. Deleting Dictionary Items

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}
del my_dict['b']  # Delete key 'b'
print(my_dict)    # Output: {'a': 1, 'c': 3}
```

# Advanced Usage

## 1. Deleting Object Attributes

```python
class MyClass:
    def __init__(self):
        self.value = 10

obj = MyClass()
print(obj.value)  # Output: 10
del obj.value
print(obj.value)  # Raises AttributeError
```

## 2. Deleting Multiple Variables

```python
x, y, z = 1, 2, 3
del x, y
print(z)  # Works
print(x)  # Raises NameError
```
