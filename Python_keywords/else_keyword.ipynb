# The else Keyword in Python
The else keyword in Python is used in conditional statements to define a block of code that runs when no previous conditions are True. It acts as a "catch-all" for situations where all other conditions fail.

# Common Uses of else
    1. With if Statements (Most Common)
        if condition:
            # Runs if condition is True
        else:
            # Runs if condition is False

    2.With Loops (for and while)
        for item in iterable:
            # Loop body
        else:
            # Runs ONLY if loop completes normally (no 'break')

    3. With try/except (Exception Handling)
        try:
            # Risky code
        except SomeError:
            # Handle error
        else:
            # Runs ONLY if no exception occurred

# Key Differences from elif
| Keyword | Purpose |
|---------|---------|
| `elif` | Checks additional conditions after an `if`. |
| `else` | Catches all remaining cases (no condition check). |

# Summary Table
| Use Case | `else` Behavior |
|----------|-----------------|
| `if` / `elif` / `else` | Runs if **all** preceding conditions are `False`. |
| `for` loop | Runs if the loop completes **without** encountering a `break` statement. |
| `while` loop | Runs if the loop completes **without** encountering a `break` statement. |
| `try` / `except` | Runs if **no exception** occurs in the `try` block. |

# Examples

## 1. else with if Statements

```python
age = 17

if age >= 18:
    print("You can vote!")
else:
    print("You cannot vote yet.")  # ✅ This runs (17 < 18)
```

## 2. Multiple Conditions (if + elif + else)

```python
score = 75

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"  # ✅ This runs
else:
    grade = "F"  # Only runs if ALL above are False

print(f"Grade: {grade}")  # Output: Grade: C
```

## 3. else with Loops (The "Loop Else" Clause)

The else block in a loop runs only if the loop completes normally (without hitting a break).

```python
#Example with for Loop

numbers = [1, 2, 3, 4, 5]
target = 10

for num in numbers:
    if num == target:
        print("Found it!")
        break
else:
    print("Target not found!")  # ✅ Runs because 'break' never happened
```

```python
# Example with break (else doesn't run)
numbers = [1, 2, 3, 4, 5]
target = 3

for num in numbers:
    if num == target:
        print("Found it!")
        break  # 💥 Breaks the loop
else:
    print("Target not found!")  # ❌ Skipped (because of 'break')
```

```python
# Example with while Loop
count = 0
while count < 5:
    print(count)
    count += 1
else:
    print("Loop completed successfully!")  # ✅ Runs after loop finishes
```

## 4. else with try/except
The else block runs only if no exceptions were raised in the try block.

```python
try:
    x = 10 / 2  # No error
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print(f"Result: {x}")  # ✅ Runs (no error occurred)
```

```python
# Example with an Exception
try:
    x = 10 / 0  # 💥 Error
except ZeroDivisionError:
    print("Cannot divide by zero!")  # ✅ Runs
else:
    print(f"Result: {x}")  # ❌ Skipped (due to error)
```
